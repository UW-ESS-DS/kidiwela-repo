```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers.convolutional import MaxPooling1D
from keras.layers import Dense, Conv2D, Flatten
#create model
model = Sequential()
#add model layers
model.add(Conv2D(60, kernel_size=(2,2), activation='relu', input_shape=(2,3000,1), padding='SAME'))
model.add(Conv2D(30, kernel_size=(2,2), activation='relu'))
model.add(Flatten())
model.add(Dense(units = 3000, activation = 'sigmoid'))
model.add(Dense(units = 1000, activation = 'sigmoid'))
model.add(Dense(units = 500, activation = 'sigmoid'))
model.add(Dense(units = 100, activation = 'sigmoid'))
model.add(Dense(units = 10, activation = 'sigmoid'))
model.add(Dense(units = 5, activation = 'sigmoid'))
model.add(Dense(units = 1, activation = 'relu'))

model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mse')
```
